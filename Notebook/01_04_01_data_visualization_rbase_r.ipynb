## Getting Started: The `plot()` Function (Scatter Plots)

The most versatile base R plotting function is `plot()`. By default, when given two numeric vectors, it creates a scatter plot.

Generate sample data:


```python
%%R
# Generate some sample data
set.seed(123) # for reproducibility
x <- 1:50
y <- x + rnorm(50, mean=0, sd=10) # y is x plus some noise
z <- x^2 - 2*x + rnorm(50, mean=0, sd=50) # another relationship
categories <- sample(c("A", "B", "C"), 50, replace=TRUE)
```

### Scatter plot

Create a first scatter plot:


```python
%%R -w 600 -h 450
# Basic Scatter Plot
plot(x, y)
```

**Adding customization to `plot()`:**


```python
%%R -w 600 -h 450
plot(x, y,
     main = "My First Scatter Plot",
     xlab = "X-axis Label",
     ylab = "Y-axis Label",
     col = "blue",
     pch = 19,
     cex = 1.2,
     xlim = c(0, 60),
     ylim = c(-20, 70)
     )
```

* **`main`**: Sets the main title of the plot.
* **`xlab`**, **`ylab`**: Set the axis labels.
* **`col`**: Point color — names (`"red"`), hex (`"#FF0000"`), or numbers.
* **`pch`**: Plotting character/symbol (`?pch`).
* **`cex`**: Point size multiplier.
* **`xlim`**, **`ylim`**: Axis limits.


### Line plots

Use `plot()` with `type = "l"`:


```python
%%R -w 600 -h 450
plot(x, z,
     type = "l",
     main = "Line Plot Example",
     xlab = "Index",
     ylab = "Value",
     col = "darkgreen",
     lwd = 2
     )
```

* **`type`**:
  * `"p"`: points (default)
  * `"l"`: lines
  * `"b"`: both points and lines
  * `"o"`: overplotted points and lines
  * `"h"`: histogram-like vertical lines
  * `"s"`: stair steps
  * `"n"`: no plotting (axes only)
* **`lwd`**: Line width.


### Histograms (`hist()`)

Histograms show the distribution of a single numeric variable.


```python
%%R -w 600 -h 450
hist(y,
     main = "Distribution of Y",
     xlab = "Y Value",
     col = "lightblue",
     border = "darkblue"
     )
```

* **`breaks`**: Bin count/boundaries, or the strings Sturges, FD, Scott, or numeric breaks.


```python
%%R -w 600 -h 450
hist(y, breaks=10, main="Histogram with 10 Bins", xlab="Y Value", col="salmon")
```

### Box plots (`boxplot()`)

Compare a numeric variable across categories (median, quartiles, outliers):


```python
%%R -w 600 -h 450
boxplot(y ~ categories,
        main = "Y by Category",
        xlab = "Category",
        ylab = "Y Value",
        col = c("coral", "lightgreen", "skyblue")
        )
```

* The formula `numeric_var ~ categorical_var` is a common pattern in R.


### Bar plots (`barplot()`)

Bar plots display counts (or summarized values) for categories. Often you count with `table()` first.


```python
%%R -w 600 -h 450
# table() builds a frequency table; barplot draws bar heights
counts <- table(categories)
barplot(counts,
        main = "Count by Category",
        xlab = "Category",
        ylab = "Count",
        col = c("coral", "lightgreen", "skyblue")
        )
```

### Adding Elements to an Existing Plot

Base graphics let you overlay lines, text, and legends on the current plot.


```python
%%R -w 600 -h 450
plot(x, y,
     main = "Scatter Plot with Added Elements",
     xlab = "X", ylab = "Y",
     pch = 16, col = "darkgray",
     cex = 1.2
     )

abline(h = mean(y), col = "red", lty = 2, lwd = 2)
abline(v = median(x), col = "purple", lty = 3, lwd = 2)

text(x = 10, y = 60, labels = "Mean Y", col = "red")
text(x = 40, y = -10, labels = "Median X", col = "purple")

legend("topleft",
       legend = c("Data Points", "Mean Y", "Median X"),
       col = c("darkgray", "red", "purple"),
       pch = c(16, NA, NA),
       lty = c(NA, 2, 3),
       lwd = c(NA, 2, 2)
       )
```

* **`abline(h=…)`**, **`abline(v=…)`**: Horizontal/vertical lines.
* **`abline(a, b)`**: Line with intercept `a` and slope `b`.
* **`text(x, y, labels)`**: Text at coordinates.
* **`legend()`**: Legend position, labels, colors, `pch`, `lty`, `lwd`.


### Multiple plots in one window (`par()`)

`par(mfrow = c(rows, cols))` arranges several plots on one device.


```python
%%R -w 960 -h 960
# 2 x 2 grid
par(mfrow = c(2, 2))

plot(x, y, main = "X vs Y", xlab = "X", ylab = "Y", col = "blue", pch = 19)
hist(y, main = "Histogram of Y", xlab = "Y Value", col = "lightblue")
boxplot(y ~ categories, main = "Y by Category", xlab = "Category", ylab = "Y Value",
        col = c("coral", "lightgreen", "skyblue"))
plot(x, z, type = "l", main = "X vs Z (Line)", xlab = "X", ylab = "Z", col = "darkgreen", lwd = 2)

par(mfrow = c(1, 1))
```

* **`par(mfrow = c(rows, cols))`**: fills **rows** first; `mfcol` fills columns first.
* **`par(mfrow = c(1, 1))`**: Reset when finished so later plots are single-panel.


## Saving plots

`png()`, `pdf()`, etc. open a graphics device; plot; then **`dev.off()`** to finalize the file:


```python
%%R
png("/content/my_scatter_plot.png", width = 800, height = 600)
plot(x, y, main = "Scatter Plot Saved as PNG", xlab = "X", ylab = "Y", col = "steelblue", pch = 16)
dev.off()

pdf("/content/my_box_plot.pdf", width = 7, height = 5)
boxplot(y ~ categories, main = "Box Plot Saved as PDF", xlab = "Category", ylab = "Y Value", col = "lightcoral")
dev.off()

cat("Files written under /content/ (download from Files pane in Colab).\n")
```

* Open a device with `png()`, `jpeg()`, `pdf()`, `svg()`, …; all drawing goes there until **`dev.off()`**.
